Functional explored:

```python
import carbonfluxtools.io_utils as cio
import carbonfluxtools.plotting as cplot
import matplotlib.pyplot as plt
import numpy as np
```

# NA functional

```python
with open('../data/na_june_functional.npy', 'rb') as f:
    h_na = np.load(f)
```

```python
plt.figure(figsize=(8, 5))
plt.plot(h_na)
plt.xlim(18000, 20000)
plt.show()
```

# Western NA Functional

```python
BASE_PATH = '/Users/mikestanley/Research/Carbon_Flux'
```

```python
# grid path
LOC_INFO = BASE_PATH + '/data/geographic_locations.json'
```

```python
# get geo locations
geo_dict = cio.read_geo_info(file_path=LOC_INFO)

# get north american grid info information
lon_pts_na, lat_pts_na = cio.get_regional_info(
    geo_dict=geo_dict, region_nm='north_america'
)
print('Longitude idxs: %s' % str(lon_pts_na))
print('Latitude idxs: %s' % str(lat_pts_na))
```

```python
# read in lon/lat arrays
with open(BASE_PATH + '/data/lon_lat_arrs/lat.npy', 'rb') as f:
    lat = np.load(f)
with open(BASE_PATH + '/data/lon_lat_arrs/lon.npy', 'rb') as f:
    lon = np.load(f)
```

#### Visualize

```python
cplot.plot_region_dots(
    lon_pts=lon_pts_na[:-6], lat_pts=lat_pts_na, extent_lst=[-128, -70, 25, 55], lon=lon, lat=lat
)
```
