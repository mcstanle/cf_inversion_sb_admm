This notebook looks at the solution vectors to the unfolding problem as solved by the ADMM optimizer. Namely, we want to see if
1. LEP and UEP solutions are *roughly* off by a sign.

There is a variety of data in `./data/*` that can be used to carry out this analysis.

```python
import matplotlib.pyplot as plt
import numpy as np
import pickle
```

# Solution signs

```python
with open('./data/unfolding_results_LEP_mu1e3.pkl', 'rb') as f:
    lep_res = pickle.load(f)
with open('./data/unfolding_results_UEP_mu1e3.pkl', 'rb') as f:
    uep_res = pickle.load(f)
```

```python
# look at the solutions
w_lep = lep_res['w_opt_output']['vectors'][-1]
w_uep = uep_res['w_opt_output']['vectors'][-1]

plt.figure(figsize=(8, 5))
plt.plot(w_lep, label='LEP')
plt.plot(w_uep, label='UEP')
plt.axhline(0, linestyle='--', color='gray')
plt.legend()
plt.show()
```
